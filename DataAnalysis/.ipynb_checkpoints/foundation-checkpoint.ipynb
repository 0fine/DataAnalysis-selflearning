# Series的用法

```python
# 创建a1，包含元素[92, 67, 70, 88, 76]及对应索引["小明", "小红", "小杰", "小丽", "小华"]
# 创建a2，包含元素[95, 85, 60, 79, 76]及对应索引["小明", "小杰", "小宇", "小娟", "小彤"]，创建后打印a1和a2
a1 = pd.Series([92, 67, 70, 88, 76], index=["小明", "小红", "小杰", "小丽", "小华"])
a2 = pd.Series([95, 85, 60, 79, 76], index=["小明", "小杰", "小宇", "小娟", "小彤"])
print(a1)
print(a2)
```

```python
# 让a1和a2的相同索引元素之间进行相加，若无法找到对应索引的值则将缺失值用0代替
a1.add(a2, fill_value=0)
```

```python
# 输出s1的统计学相关数据，包括最大值、最小值、平均值、标准差
a1.describe()
```

```python
# 把a1更新为所有数字加5后的结果，并进行输出
a1 + 5   # ————这样写只会输出结果，但不会改变a1
```

```python
# 根据规则，输出一个新的Series，包含a1里各个学生成绩所对应的等级
def get_grade_from_score(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return '不及格'
```

```python
get_grade_from_score(55)
```

```python
a1.apply(get_grade_from_score)    # apply相当于一个高阶函数，在apply里调用函数
```

# DataFrame的用法

```python
# DataFrame可以看成由Series组成的字典
# df[] 为列的索引
# df.loc[]   df.iloc[]  为行的索引
```

```python
# 创建一个叫name的Series，里面包含字符串元素 "小陈", '小李', '小王', '小张', '小赵', '小周' ，标签为 '001', '002', '003', '004', '005', '006'
# 创建一个叫gender的Series，里面包含字符串元素 '女', '女', '男', '男', '女', '男' ， 标签为 '006', '005', '004', '003', '002', '001'
# 创建一个叫height的Series，里面包含浮点数元素 172.5, 168.0, 178.2, 181.3, 161.7 ， 标签为 '001', '002', '003', '004', '005'
```

```python
name = pd.Series(["小陈", '小李', '小王', '小张', '小赵', '小周'], index=['001', '002', '003', '004', '005', '006'])
gender = pd.Series(['女', '女', '男', '男', '女', '男'], index=['006', '005', '004', '003', '002', '001'])
height = pd.Series([172.5, 168.0, 178.2, 181.3, 161.7], index=['001', '002', '003', '004', '005'])
```

```python
# 创建一个叫students的DataFrame，包含列名姓名、性别、身高，数据分别对应创建好的name、gender和height，这三个Series，输出创建好的students
```

```python
students = pd.DataFrame({'姓名': name, '性别': gender, '身高': height})    # 三个Series，根据索引对齐，进行合并
students
```

```python
# 创建一个名为students的DataFrame，包含数据如下： 
# {"001": {"姓名": "小陈", "考试1": 85, "考试2": 95, "考试3": 92},  "002": {"姓名": "小李", "考试1": 91, "考试2": 92, "考试3": 94},  
#  "003": {"姓名": "小王", "考试1": 86, "考试2": 81, "考试3": 89},  "004": {"姓名": "小张", "考试1": 79, "考试2": 89, "考试3": 95},
#  "005": {"姓名": "小赵", "考试1": 96, "考试2": 91, "考试3": 91},  "006": {"姓名": "小周", "考试1": 81, "考试2": 89, "考试3": 92}  }
# 要求索引为学号（001、002等），列名为姓名、考试1、考试2、考试3，并输出创建好的DataFrame

students = pd.DataFrame({"001": {"姓名": "小陈", "考试1": 85, "考试2": 95, "考试3": 92},  
                             "002": {"姓名": "小李", "考试1": 91, "考试2": 92, "考试3": 94},  
                             "003": {"姓名": "小王", "考试1": 86, "考试2": 81, "考试3": 89},  
                             "004": {"姓名": "小张", "考试1": 79, "考试2": 89, "考试3": 95}, 
                             "005": {"姓名": "小赵", "考试1": 96, "考试2": 91, "考试3": 91},  
                             "006": {"姓名": "小周", "考试1": 81, "考试2": 89, "考试3": 92}  }).T
students
```

```python
# 创建一个students_average的DataFrame，索引为学号，包含姓名列（列名为“姓名”），以及计算出三次考试平均值后的结果列（列名为“平均分”）
average = students.loc[:, "考试1":"考试3"].mean(axis = 1)
name = students["姓名"]
students_average = pd.DataFrame({"姓名": name, "平均分": average})
students_average
```

```python
# 输出每次考试第二高的分数组成的Series，可以尝试结合匿名函数
students.loc[:, "考试1":"考试3"].apply(lambda x : np.sort(x)[-2])
```

```python
def get_grade(score):
    if score >= 95:
        return "A+"
    elif score >= 90:
        return "A"
    elif score >= 85:
        return "B+"
    elif score >= 80:
        return "B"
    elif score >= 75:
        return "C+"
    else:
        return "C"
```

```python
students.loc[:, "考试1" : "考试2"] .applymap(get_grade)
```

```python
# 展示students每个考试分数数据列的统计信息，包括个数、平均值、标准差、四分位数
students.describe()
```
